## 레이블 생성 함수 정의

- 레이블 생성 함수는 연월을 인자로 받아서 고객 아이디별로 총 구매액이 인자로 받는 구매액 임계값을 넘는지 여부를 바이너리 레이블로 생성하는 함수

```python
TOTAL_THRES = 300

'''
    입력인자로 받는 year_month에 대해 고객 ID별로 총 구매액이
    구매액 임계값을 넘는지 여부의 binary label을 생성하는 함수
'''
def generate_label( #각 라벨에 대한 타입지정
    df: pd.DataFrame,
    year_month: str,
    total_thres: int = TOTAL_THRES,
    print_log: bool = False
):
    df = df.copy()#해당 데이터프레임에서 df만 복사

    # year_month에 해당하는 label 데이터 생성
    df['year_month'] = df['order_date'].dt.strftime('%Y-%m') #orderdate를 연도_달 형태로 바꿔서 표시
    df.reset_index(drop=True, inplace=True) #원래 열을 바뀐 형태로 교체

    # year_month 이전 월의 고객 ID 추출
    cust = df[df['year_month']<year_month]['customer_id'].unique() # 겹치지 않게, year_month보다 작은 고객id 추출
    # year_month에 해당하는 데이터 선택
    df = df[df['year_month']==year_month] #입력받은 달과 같은 데이터열만 추출

    # label 데이터프레임 생성
    label = pd.DataFrame({'customer_id':cust}) #year_month가 작은 고객들로 새로 데이터 프레임 생성
    label['year_month'] = year_month  # 해당 데이터 프레임의 year_month를 입력받은 값들로 채워넣음

    # year_month에 해당하는 고객 ID의 구매액의 합 계산
    grped = df.groupby(['customer_id','year_month'], as_index=False)[['total']].sum() # 고객id과 날짜별로 묶어서 total의 총합 구함

    # label 데이터프레임과 merge하고 구매액 임계값을 넘었는지 여부로 label 생성
    label = label.merge(grped, on=['customer_id','year_month'], how='left') #고객 id와 날짜별로 묶인 total값을 추가?
    label['total'].fillna(0.0, inplace=True) # total의 값을 모두 nan으로 초기화?
    label['label'] = (label['total'] > total_thres).astype(int) # total_thres 넘으면 1, 아니면 0

    # 고객 ID로 정렬
    label = label.sort_values('customer_id').reset_index(drop=True) #id별로 재정렬하고 인덱스 재설정
    if print_log: print(f'{year_month} - final label shape: {label.shape}')

    return label
```

## 평가지표 출력 함수 정의

```python
'''
    평가지표를 출력하는 함수
'''
def print_score(label: List[float], pred: List[float], prob_thres: float = 0.5): #print_score의 형태 지정
    print('Precision: {:.5f}'.format(precision_score(label, pred>prob_thres))) # sklearn 라이브러리의 정밀도 함수
    print('Recall: {:.5f}'.format(recall_score(label, pred>prob_thres))) # 재현율 함수
    print('F1 Score: {:.5f}'.format(f1_score(label, pred>prob_thres))) # f1스코어
    print('ROC AUC Score: {:.5f}'.format(roc_auc_score(label, pred))) # auc 스코어, 모델의 분류성능
```

## 베이스라인 모델 함수
- 머신러닝 모델 없이 입력 인자로 받은 연월의 이전 달 총 구매액을 구매 확률로 예측하는 베이스라인 모델 함수

```python
'''
    머신러닝 모델 없이 입력인자으로 받는 year_month의 이전 달 총 구매액을 구매 확률로 예측하는 베이스라인 모델
'''
def baseline_no_ml(df: pd.DataFrame, year_month: str, total_thres=TOTAL_THRES):
    # year_month에 해당하는 label 데이터 생성
    month = generate_label(df, year_month)

    # year_month 이전 월 계산
    d = datetime.datetime.strptime(year_month, "%Y-%m")
    prev_d = d - dateutil.relativedelta.relativedelta(months=1)
    prev_d = prev_d.strftime('%Y-%m')

    # 이전 월에 해당하는 label 데이터 생성
    previous_month = generate_label(df, prev_d)

    # merge하기 위해 컬럼명 변경
    previous_month = previous_month.rename(columns = {'total': 'previous_total'})

    month = month.merge(previous_month[['customer_id', 'previous_total']], on = 'customer_id', how = 'left')

    # 거래내역이 없는 고객의 구매액을 0으로 채움
    month['previous_total'] = month['previous_total'].fillna(0)
    # 이전 월의 총 구매액을 구매액 임계값으로 나눠서 예측 확률로 계산
    month['probability'] = month['previous_total'] / total_thres

    # 이전 월 총 구매액이 구매액 임계값을 넘어서 1보다 클 경우 예측 확률을 1로 변경
    month.loc[month['probability'] > 1, 'probability'] = 1

    # 이전 월 총 구매액이 마이너스(주문 환불)일 경우 예측 확률을 0으로 변경
    month.loc[month['probability'] < 0, 'probability'] = 0

    return month['probability']
```
