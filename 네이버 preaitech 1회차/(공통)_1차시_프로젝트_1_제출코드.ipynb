다음 지시문의 코드를 작성하고 결과를 출력해보세요.

```python
!pip install numpy
import numpy as np

# 주어진 리스트
data = [10, 25, 40, 55, 70]
```

```python
## 코드시작 ##
# 1. numpy 배열로 변환하여 출력하세요.
np_array=np.array(data)
print(np_array)

# 2. 리스트의 모든 원소의 합을 계산하여 출력하세요.
print(sum(np_array))

# 3. 리스트의 모든 원소의 평균을 계산하여 출력하세요.
print(np.mean(np_array))

# 4. 리스트의 최댓값과 최솟값을 출력하세요.
print(max(np_array))
print(min(np_array))

# 5. 각 원소를 10으로 나누는 연산을 수행하고 결과를 출력하세요.
print(np_array/10)

# 6. 각 원소를 3으로 거듭제곱하는 연산을 수행하고 결과를 출력하세요.
print(np.power(np_array,3))

## 코드종료 ##
```
